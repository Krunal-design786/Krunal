```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_excel('/kaggle/input/shopping-trends/SHOPPING_TRENDS.xlsx')
```

```python
df.drop(['Customer name','Expiry date','Manufacturing date'], axis=1, inplace=True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Previous Purchases'] = df['Previous Purchases'].astype('int')
```

```python
df['Previous Purchases'].dtypes
```

```python
df.describe()
```

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Purchase Amount (USD)'].sum().sort_values(by='Purchase Amount (USD)',ascending=False)

ax = sns.barplot(x = 'Gender',y = 'Purchase Amount (USD)',data = sales_gen)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_location = df.groupby(['Location'],as_index=False)['Purchase Amount (USD)'].sum().sort_values(by='Purchase Amount (USD)',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
ax = sns.barplot(x = 'Location',y = 'Purchase Amount (USD)',data = sales_location)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df,x = 'Frequency of Purchases')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df,x = 'Review Rating')

for bars in ax.containers:
    ax.bar_label(bars)
```
